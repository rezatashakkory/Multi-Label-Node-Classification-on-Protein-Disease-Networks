## 1. Setup and Data Loading

```python
import torch
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from scipy import stats
from scipy.sparse import csr_matrix
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

print("âœ“ Libraries loaded")
```

```python
# Load data
data_dir = '../data/'

edge_index = torch.load(data_dir + 'edge_index.pt')
node_features = torch.load(data_dir + 'node_features.pt')
y = torch.load(data_dir + 'y.pt')
train_idx = torch.load(data_dir + 'train_idx.pt')
test_idx = torch.load(data_dir + 'test_idx.pt')

num_nodes = node_features.shape[0]
num_edges = edge_index.shape[1]
num_features = node_features.shape[1]
num_labels = y.shape[1]

print("="*80)
print("DATA LOADED")
print("="*80)
print(f"Nodes: {num_nodes:,}")
print(f"Edges: {num_edges:,}")
print(f"Features: {num_features}")
print(f"Labels (diseases): {num_labels}")
print(f"Train nodes: {len(train_idx):,}")
print(f"Test nodes: {len(test_idx):,}")
print("="*80)
```

## 2. Basic Graph Statistics

```python
# Create NetworkX graph for analysis
G = nx.Graph()
edges = edge_index.t().numpy()
G.add_edges_from(edges)

print("\n" + "="*80)
print("BASIC GRAPH PROPERTIES")
print("="*80)

# Basic statistics
print(f"\nðŸ“Š Connectivity:")
print(f"  Is connected: {nx.is_connected(G)}")
num_components = nx.number_connected_components(G)
print(f"  Number of connected components: {num_components}")

if num_components > 1:
    largest_cc = max(nx.connected_components(G), key=len)
    print(f"  Largest component size: {len(largest_cc):,} nodes ({len(largest_cc)/num_nodes*100:.1f}%)")

# Density
density = nx.density(G)
print(f"\nðŸ“Š Density: {density:.6f}")
print(f"  (Current edges: {num_edges:,} / Possible: {num_nodes*(num_nodes-1)//2:,})")

# Degree statistics
degrees = dict(G.degree())
degree_vals = list(degrees.values())
print(f"\nðŸ“Š Degree Statistics:")
print(f"  Min degree: {min(degree_vals)}")
print(f"  Max degree: {max(degree_vals)}")
print(f"  Mean degree: {np.mean(degree_vals):.2f}")
print(f"  Median degree: {np.median(degree_vals):.2f}")
print(f"  Std degree: {np.std(degree_vals):.2f}")

# Top hub genes
top_hubs = sorted(degrees.items(), key=lambda x: x[1], reverse=True)[:10]
print(f"\nðŸŽ¯ Top 10 Hub Genes (highest degree):")
for i, (node, deg) in enumerate(top_hubs, 1):
    print(f"  {i:2d}. Node {node:5d}: degree = {deg:4d}")

print("\n" + "="*80)
```

## 3. Degree Distribution Analysis

```python
# Analyze degree distribution
degree_counts = Counter(degree_vals)
degrees_sorted = sorted(degree_counts.keys())
counts = [degree_counts[d] for d in degrees_sorted]

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Degree distribution (linear scale)
ax = axes[0, 0]
ax.bar(degrees_sorted, counts, alpha=0.7, color='steelblue', edgecolor='black')
ax.set_xlabel('Degree', fontsize=12)
ax.set_ylabel('Number of Nodes', fontsize=12)
ax.set_title('Degree Distribution (Linear Scale)', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)

# Plot 2: Log-log scale (check for power law)
ax = axes[0, 1]
ax.scatter(degrees_sorted, counts, alpha=0.6, color='coral', s=30)
ax.set_xlabel('Degree (log scale)', fontsize=12)
ax.set_ylabel('Number of Nodes (log scale)', fontsize=12)
ax.set_title('Degree Distribution (Log-Log Scale)', fontsize=14, fontweight='bold')
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid(True, alpha=0.3)

# Check for power law
# Fit: log(count) = a + b*log(degree)
log_degrees = np.log10([d for d in degrees_sorted if d > 0])
log_counts = np.log10([degree_counts[d] for d in degrees_sorted if d > 0])
slope, intercept, r_value, _, _ = stats.linregress(log_degrees, log_counts)
fit_line = slope * log_degrees + intercept
ax.plot(10**log_degrees, 10**fit_line, 'r--', linewidth=2, 
        label=f'Power law fit: Î³={-slope:.2f}\n$R^2$={r_value**2:.3f}')
ax.legend(fontsize=10)

# Plot 3: CCDF (Complementary Cumulative Distribution)
ax = axes[1, 0]
degrees_array = np.array(degree_vals)
unique_degrees = np.unique(degrees_array)
ccdf = [np.sum(degrees_array >= d) / len(degrees_array) for d in unique_degrees]
ax.plot(unique_degrees, ccdf, marker='o', linestyle='-', color='green', alpha=0.6)
ax.set_xlabel('Degree', fontsize=12)
ax.set_ylabel('P(Degree â‰¥ k)', fontsize=12)
ax.set_title('Complementary Cumulative Distribution (CCDF)', fontsize=14, fontweight='bold')
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid(True, alpha=0.3)

# Plot 4: Box plot and violin plot
ax = axes[1, 1]
violin_parts = ax.violinplot([degree_vals], positions=[0], widths=0.7,
                             showmeans=True, showmedians=True)
ax.set_ylabel('Degree', fontsize=12)
ax.set_title('Degree Distribution (Violin Plot)', fontsize=14, fontweight='bold')
ax.set_xticks([0])
ax.set_xticklabels(['All Nodes'])
ax.grid(True, alpha=0.3, axis='y')

# Add statistics text
stats_text = f"Mean: {np.mean(degree_vals):.1f}\n"
stats_text += f"Median: {np.median(degree_vals):.1f}\n"
stats_text += f"Q1: {np.percentile(degree_vals, 25):.1f}\n"
stats_text += f"Q3: {np.percentile(degree_vals, 75):.1f}"
ax.text(0.5, 0.95, stats_text, transform=ax.transAxes,
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
        fontsize=10)

plt.tight_layout()
plt.savefig('../data/degree_distribution_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nðŸ’¡ Interpretation:")
if -slope > 2:
    print(f"  âœ“ Power-law exponent Î³ = {-slope:.2f} indicates scale-free network")
    print(f"  âœ“ Few hub nodes with very high degree, many nodes with low degree")
    print(f"  âœ“ Typical of biological networks (protein-protein interactions)")
else:
    print(f"  âœ“ Degree distribution deviates from pure scale-free (Î³ = {-slope:.2f})")

print(f"  âœ“ RÂ² = {r_value**2:.3f} shows {'good' if r_value**2 > 0.8 else 'moderate'} fit to power law")
```

## 4. Label-Specific Analysis & Homophily

**Key Insight**: Per-label homophily explains why some diseases are easier to predict

```python
print("\n" + "="*80)
print("LABEL-SPECIFIC ANALYSIS (Training Data)")
print("="*80)

# Focus on training data
train_labels = y[train_idx].numpy()
train_edges_mask = np.isin(edge_index[0].numpy(), train_idx.numpy()) & \
                   np.isin(edge_index[1].numpy(), train_idx.numpy())
train_edges = edge_index[:, train_edges_mask]

# Convert train_idx to local indices
train_idx_np = train_idx.numpy()
node_to_local = {node: i for i, node in enumerate(train_idx_np)}

# Calculate per-label statistics
label_stats = []

for label_idx in range(num_labels):
    # Positive nodes for this label
    pos_mask = train_labels[:, label_idx] == 1
    num_positive = pos_mask.sum()
    
    if num_positive == 0:
        continue
    
    # Degrees of positive nodes
    pos_nodes_global = train_idx_np[pos_mask]
    pos_degrees = [degrees[int(node)] for node in pos_nodes_global]
    avg_degree = np.mean(pos_degrees)
    
    # Homophily: fraction of edges between positive nodes
    # Count edges where both endpoints are positive
    homophily_edges = 0
    total_edges = 0
    
    for i in range(train_edges.shape[1]):
        src, dst = train_edges[0, i].item(), train_edges[1, i].item()
        if src in node_to_local and dst in node_to_local:
            src_local = node_to_local[src]
            dst_local = node_to_local[dst]
            
            # At least one endpoint is positive
            if pos_mask[src_local] or pos_mask[dst_local]:
                total_edges += 1
                # Both endpoints are positive
                if pos_mask[src_local] and pos_mask[dst_local]:
                    homophily_edges += 1
    
    homophily = homophily_edges / total_edges if total_edges > 0 else 0
    
    label_stats.append({
        'label_id': label_idx,
        'num_positive': num_positive,
        'avg_degree': avg_degree,
        'homophily': homophily
    })
    
    if label_idx < 5:  # Print first 5 for example
        print(f"\nLabel {label_idx}:")
        print(f"  Positive nodes: {num_positive}")
        print(f"  Avg degree: {avg_degree:.2f}")
        print(f"  Homophily: {homophily:.4f}")

# Create DataFrame
label_df = pd.DataFrame(label_stats)

print(f"\n\nðŸ“Š Label Statistics Summary:")
print(f"  Total labels analyzed: {len(label_df)}")
print(f"  Positive samples range: [{label_df['num_positive'].min()}, {label_df['num_positive'].max()}]")
print(f"  Avg degree range: [{label_df['avg_degree'].min():.2f}, {label_df['avg_degree'].max():.2f}]")
print(f"  Homophily range: [{label_df['homophily'].min():.4f}, {label_df['homophily'].max():.4f}]")
print(f"  Mean homophily: {label_df['homophily'].mean():.4f}")

print("\n" + "="*80)
```

```python
# Visualize label statistics
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Number of positive samples distribution
ax = axes[0, 0]
ax.hist(label_df['num_positive'], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
ax.set_xlabel('Number of Positive Samples', fontsize=12)
ax.set_ylabel('Number of Labels', fontsize=12)
ax.set_title('Distribution of Positive Samples per Label', fontsize=14, fontweight='bold')
ax.axvline(label_df['num_positive'].median(), color='red', linestyle='--', 
           label=f'Median: {label_df["num_positive"].median():.0f}')
ax.legend()
ax.grid(True, alpha=0.3)

# Plot 2: Homophily distribution
ax = axes[0, 1]
ax.hist(label_df['homophily'], bins=50, alpha=0.7, color='lightcoral', edgecolor='black')
ax.set_xlabel('Homophily Score', fontsize=12)
ax.set_ylabel('Number of Labels', fontsize=12)
ax.set_title('Distribution of Label Homophily', fontsize=14, fontweight='bold')
ax.axvline(label_df['homophily'].mean(), color='darkred', linestyle='--', 
           label=f'Mean: {label_df["homophily"].mean():.4f}')
ax.legend()
ax.grid(True, alpha=0.3)

# Plot 3: Homophily vs Positive samples
ax = axes[1, 0]
scatter = ax.scatter(label_df['num_positive'], label_df['homophily'], 
                    alpha=0.5, c=label_df['avg_degree'], cmap='viridis', s=50)
ax.set_xlabel('Number of Positive Samples', fontsize=12)
ax.set_ylabel('Homophily', fontsize=12)
ax.set_title('Homophily vs Positive Samples\n(color = avg degree)', 
             fontsize=14, fontweight='bold')
plt.colorbar(scatter, ax=ax, label='Avg Degree')
ax.grid(True, alpha=0.3)

# Add correlation
corr = label_df[['num_positive', 'homophily']].corr().iloc[0, 1]
ax.text(0.05, 0.95, f'Correlation: {corr:.3f}', transform=ax.transAxes,
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.7))

# Plot 4: Average degree vs homophily
ax = axes[1, 1]
scatter = ax.scatter(label_df['avg_degree'], label_df['homophily'], 
                    alpha=0.5, c=label_df['num_positive'], cmap='plasma', s=50)
ax.set_xlabel('Average Degree of Positive Nodes', fontsize=12)
ax.set_ylabel('Homophily', fontsize=12)
ax.set_title('Degree vs Homophily\n(color = num positive)', 
             fontsize=14, fontweight='bold')
plt.colorbar(scatter, ax=ax, label='Num Positive')
ax.grid(True, alpha=0.3)

# Add correlation
corr = label_df[['avg_degree', 'homophily']].corr().iloc[0, 1]
ax.text(0.05, 0.95, f'Correlation: {corr:.3f}', transform=ax.transAxes,
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.7))

plt.tight_layout()
plt.savefig('../data/label_homophily_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nðŸ’¡ Key Insights:")
print(f"  âœ“ Mean homophily: {label_df['homophily'].mean():.4f}")
if label_df['homophily'].mean() > 0.5:
    print("  âœ“ HIGH homophily â†’ GNNs should work well (neighbors share labels)")
elif label_df['homophily'].mean() > 0.3:
    print("  âœ“ MODERATE homophily â†’ GNNs + feature-based models recommended")
else:
    print("  âœ“ LOW homophily â†’ Feature-based models may outperform GNNs")

print(f"  âœ“ Labels with high homophily are easier to predict")
print(f"  âœ“ Hub genes (high degree) may be more disease-associated")
```

## 5. Centrality Measures

```python
print("\n" + "="*80)
print("CENTRALITY ANALYSIS (Computing on largest component)")
print("="*80)

# Work with largest connected component for efficiency
if num_components > 1:
    largest_cc = max(nx.connected_components(G), key=len)
    G_main = G.subgraph(largest_cc).copy()
    print(f"Analyzing largest component: {len(G_main.nodes()):,} nodes")
else:
    G_main = G

# Sample for expensive computations if needed
if len(G_main.nodes()) > 5000:
    print("âš ï¸  Large graph detected - computing approximate centralities on sample")
    sample_nodes = list(G_main.nodes())[:5000]
    G_sample = G_main.subgraph(sample_nodes).copy()
else:
    G_sample = G_main

# Compute centralities
print("\nComputing centrality measures...")

# Degree centrality (fast)
print("  1/5 Degree centrality...")
degree_cent = nx.degree_centrality(G_sample)

# Betweenness centrality (slow - use sampling)
print("  2/5 Betweenness centrality (approximate)...")
betweenness_cent = nx.betweenness_centrality(G_sample, k=min(500, len(G_sample.nodes())))

# Closeness centrality (moderate)
print("  3/5 Closeness centrality...")
closeness_cent = nx.closeness_centrality(G_sample)

# Eigenvector centrality (moderate)
print("  4/5 Eigenvector centrality...")
try:
    eigenvector_cent = nx.eigenvector_centrality(G_sample, max_iter=1000)
except:
    print("    (Failed - using degree as proxy)")
    eigenvector_cent = degree_cent

# PageRank (fast)
print("  5/5 PageRank...")
pagerank_cent = nx.pagerank(G_sample)

print("\nâœ“ Centrality measures computed!")

# Find top nodes
print("\nðŸŽ¯ Top 10 Most Central Genes:")
print("\nBy Degree Centrality:")
top_degree = sorted(degree_cent.items(), key=lambda x: x[1], reverse=True)[:10]
for i, (node, cent) in enumerate(top_degree, 1):
    print(f"  {i:2d}. Node {node:5d}: {cent:.6f}")

print("\nBy Betweenness Centrality:")
top_between = sorted(betweenness_cent.items(), key=lambda x: x[1], reverse=True)[:10]
for i, (node, cent) in enumerate(top_between, 1):
    print(f"  {i:2d}. Node {node:5d}: {cent:.6f}")

print("\nBy PageRank:")
top_pr = sorted(pagerank_cent.items(), key=lambda x: x[1], reverse=True)[:10]
for i, (node, cent) in enumerate(top_pr, 1):
    print(f"  {i:2d}. Node {node:5d}: {cent:.6f}")

print("\n" + "="*80)
```

```python
# Visualize centrality distributions
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

centralities = {
    'Degree': degree_cent,
    'Betweenness': betweenness_cent,
    'Closeness': closeness_cent,
    'Eigenvector': eigenvector_cent,
    'PageRank': pagerank_cent
}

for idx, (name, cent_dict) in enumerate(centralities.items()):
    row = idx // 3
    col = idx % 3
    ax = axes[row, col]
    
    values = list(cent_dict.values())
    ax.hist(values, bins=50, alpha=0.7, color=f'C{idx}', edgecolor='black')
    ax.set_xlabel(f'{name} Centrality', fontsize=11)
    ax.set_ylabel('Number of Nodes', fontsize=11)
    ax.set_title(f'{name} Centrality Distribution', fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.3)
    
    # Add statistics
    stats_text = f"Mean: {np.mean(values):.6f}\n"
    stats_text += f"Max: {np.max(values):.6f}"
    ax.text(0.98, 0.98, stats_text, transform=ax.transAxes,
            verticalalignment='top', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
            fontsize=9)

# Remove empty subplot
fig.delaxes(axes[1, 2])

plt.tight_layout()
plt.savefig('../data/centrality_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 6. Feature-Label Correlation Analysis

```python
print("\n" + "="*80)
print("FEATURE-LABEL CORRELATION ANALYSIS")
print("="*80)

# Clean features (impute NaN)
features_clean = node_features.clone()
for i in range(features_clean.shape[1]):
    col = features_clean[:, i]
    if torch.isnan(col).any():
        median_val = col[~torch.isnan(col)].median()
        features_clean[:, i] = torch.where(torch.isnan(col), median_val, col)

# Focus on training data
train_features = features_clean[train_idx].numpy()
train_labels = y[train_idx].numpy()

# Compute correlation between each feature and each label
print("\nComputing feature-label correlations...")
correlations = np.zeros((num_features, num_labels))

for feat_idx in range(num_features):
    for label_idx in range(num_labels):
        if train_labels[:, label_idx].sum() > 0:  # Skip labels with no positives
            try:
                corr, _ = stats.pearsonr(train_features[:, feat_idx], train_labels[:, label_idx])
                correlations[feat_idx, label_idx] = abs(corr)  # Absolute correlation
            except:
                correlations[feat_idx, label_idx] = 0.0  # If correlation fails, set to 0

# Find most predictive features
feature_importance = np.nanmean(correlations, axis=1)  # Average across all labels (ignore NaN)
top_features = np.argsort(feature_importance)[::-1][:10]

print("\nðŸŽ¯ Top 10 Most Predictive Features (by avg correlation):")
for i, feat_idx in enumerate(top_features, 1):
    print(f"  {i:2d}. Feature {feat_idx:2d}: avg correlation = {feature_importance[feat_idx]:.4f}")

# Find most predictable labels
label_predictability = np.nanmax(correlations, axis=0)  # Max correlation with any feature
label_predictability = np.nan_to_num(label_predictability, nan=0.0)  # Replace NaN with 0

# Get valid top labels (non-zero predictability)
valid_mask = label_predictability > 0
valid_labels = np.where(valid_mask)[0]
if len(valid_labels) > 0:
    top_label_indices = np.argsort(label_predictability[valid_labels])[::-1][:10]
    top_labels = valid_labels[top_label_indices]
    
    print("\nðŸŽ¯ Top 10 Most Predictable Labels (by max feature correlation):")
    for i, label_idx in enumerate(top_labels, 1):
        print(f"  {i:2d}. Label {label_idx:3d}: max correlation = {label_predictability[label_idx]:.4f}")
else:
    print("\nâš ï¸ No labels with valid correlations found")

print("\n" + "="*80)
```

```python
# Visualize correlation matrix (sample)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Feature importance
ax = axes[0]
ax.bar(range(num_features), sorted(feature_importance, reverse=True), 
       alpha=0.7, color='steelblue', edgecolor='black')
ax.set_xlabel('Feature (sorted by importance)', fontsize=12)
ax.set_ylabel('Average Absolute Correlation', fontsize=12)
ax.set_title('Feature Importance for Label Prediction', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

# Plot 2: Label predictability (filter out zeros for better visualization)
ax = axes[1]
# Only plot non-zero predictabilities
valid_predictability = label_predictability[label_predictability > 0]
if len(valid_predictability) > 0:
    ax.hist(valid_predictability, bins=min(50, len(valid_predictability)), 
            alpha=0.7, color='coral', edgecolor='black')
    ax.axvline(valid_predictability.mean(), color='darkred', linestyle='--',
               label=f'Mean: {valid_predictability.mean():.4f}')
    ax.legend()
else:
    ax.text(0.5, 0.5, 'No valid correlations', 
            ha='center', va='center', transform=ax.transAxes, fontsize=14)
ax.set_xlabel('Max Feature Correlation', fontsize=12)
ax.set_ylabel('Number of Labels', fontsize=12)
ax.set_title('Distribution of Label Predictability', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../data/feature_label_correlation.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nðŸ’¡ Insights:")
print(f"  âœ“ Top feature (#{top_features[0]}) correlates {feature_importance[top_features[0]]:.4f} with labels")
print(f"  âœ“ Most predictable label shows {label_predictability.max():.4f} correlation")
print(f"  âœ“ {(label_predictability > 0.1).sum()}/{num_labels} labels have strong feature correlation (>0.1)")
print(f"  âœ“ Valid labels with correlations: {(label_predictability > 0).sum()}/{num_labels}")
```

## 7. Graph Visualization (Subgraph Sample)

```python
print("\n" + "="*80)
print("GRAPH VISUALIZATION (Subgraph Sample)")
print("="*80)

# Sample a subgraph for visualization
# Strategy: Take top hub nodes and their neighbors
num_seed_nodes = 50
seed_nodes = [node for node, _ in top_degree[:num_seed_nodes]]

# Get neighborhood
subgraph_nodes = set(seed_nodes)
for node in seed_nodes:
    if node in G:
        subgraph_nodes.update(G.neighbors(node))

subgraph_nodes = list(subgraph_nodes)[:500]  # Limit to 500 nodes
G_sub = G.subgraph(subgraph_nodes).copy()

print(f"\nSubgraph for visualization:")
print(f"  Nodes: {len(G_sub.nodes())}")
print(f"  Edges: {len(G_sub.edges())}")

# Prepare node colors based on degree
node_degrees = dict(G_sub.degree())
node_colors = [node_degrees[node] for node in G_sub.nodes()]
node_sizes = [30 + node_degrees[node] * 2 for node in G_sub.nodes()]

# Layout
print("\nComputing layout (this may take a minute)...")
pos = nx.spring_layout(G_sub, k=0.5, iterations=50, seed=42)

# Plot
fig, ax = plt.subplots(figsize=(16, 16))

# Draw edges
nx.draw_networkx_edges(G_sub, pos, alpha=0.2, width=0.5, ax=ax)

# Draw nodes
nodes = nx.draw_networkx_nodes(
    G_sub, pos, 
    node_color=node_colors,
    node_size=node_sizes,
    cmap='YlOrRd',
    alpha=0.8,
    ax=ax
)

# Colorbar
plt.colorbar(nodes, label='Node Degree', ax=ax, shrink=0.8)

# Highlight top hubs
hub_nodes = [n for n in seed_nodes[:10] if n in G_sub]
hub_pos = {n: pos[n] for n in hub_nodes if n in pos}
if hub_pos:
    nx.draw_networkx_nodes(
        G_sub, hub_pos,
        nodelist=hub_nodes,
        node_color='darkred',
        node_size=300,
        node_shape='*',
        alpha=1.0,
        label='Top 10 Hubs',
        ax=ax
    )

ax.set_title('Gene Interaction Network (Subgraph Sample)\nNode size & color = degree', 
             fontsize=16, fontweight='bold', pad=20)
ax.legend(loc='upper right', fontsize=12)
ax.axis('off')

plt.tight_layout()
plt.savefig('../data/graph_visualization.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nâœ“ Visualization saved!")
print("\nðŸ’¡ Observation:")
print("  âœ“ Hub nodes (red stars) are central to the network")
print("  âœ“ Network shows clustering (modules of related genes)")
print("  âœ“ This structure is exploited by GNNs and label propagation")
```

## 8. Feature Embedding Visualization (PCA & t-SNE)

```python
print("\n" + "="*80)
print("FEATURE EMBEDDING SPACE VISUALIZATION")
print("="*80)

# Prepare features
features_np = features_clean[train_idx].numpy()
labels_np = y[train_idx].numpy()

# Use first label for coloring
label_0 = labels_np[:, 0]

# Standardize
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_np)

# PCA
print("\nComputing PCA...")
pca = PCA(n_components=2, random_state=42)
features_pca = pca.fit_transform(features_scaled)
print(f"  Explained variance: {pca.explained_variance_ratio_.sum():.2%}")

# t-SNE (on subsample for speed)
print("\nComputing t-SNE (on 2000 samples)...")
sample_size = min(2000, len(features_scaled))
sample_indices = np.random.choice(len(features_scaled), sample_size, replace=False)
features_sample = features_scaled[sample_indices]
labels_sample = label_0[sample_indices]

tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
features_tsne = tsne.fit_transform(features_sample)

print("\nâœ“ Embeddings computed!")
```

```python
# Visualize
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# PCA plot
ax = axes[0]
scatter = ax.scatter(
    features_pca[:, 0], features_pca[:, 1],
    c=label_0, cmap='RdYlBu_r', alpha=0.5, s=20
)
ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} var)', fontsize=12)
ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} var)', fontsize=12)
ax.set_title('PCA of Node Features\n(colored by Label 0)', fontsize=14, fontweight='bold')
plt.colorbar(scatter, ax=ax, label='Label 0')
ax.grid(True, alpha=0.3)

# t-SNE plot
ax = axes[1]
scatter = ax.scatter(
    features_tsne[:, 0], features_tsne[:, 1],
    c=labels_sample, cmap='RdYlBu_r', alpha=0.5, s=20
)
ax.set_xlabel('t-SNE Dimension 1', fontsize=12)
ax.set_ylabel('t-SNE Dimension 2', fontsize=12)
ax.set_title('t-SNE of Node Features\n(colored by Label 0, n=2000)', fontsize=14, fontweight='bold')
plt.colorbar(scatter, ax=ax, label='Label 0')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../data/embedding_visualization.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nðŸ’¡ Interpretation:")
print("  âœ“ If positive/negative nodes cluster â†’ features are informative")
print("  âœ“ Mixed distribution â†’ graph structure is crucial")
print("  âœ“ t-SNE reveals non-linear patterns better than PCA")
```

## 10. Export Key Metrics for Ensemble Design

```python
# Collect all key metrics for decision-making
print("\n" + "="*80)
print("KEY METRICS SUMMARY FOR ENSEMBLE DESIGN")
print("="*80)

metrics = {
    'Graph Structure': {
        'Total Nodes': num_nodes,
        'Total Edges': num_edges,
        'Mean Degree': np.mean(degree_vals),
        'Max Degree': max(degree_vals),
        'Power-law Exponent (Î³)': -slope,
        'Power-law RÂ²': r_value**2,
        'Density': density,
    },
    'Label Properties': {
        'Total Labels': num_labels,
        'Mean Homophily': label_df['homophily'].mean(),
        'Median Homophily': label_df['homophily'].median(),
        'Homophily Std': label_df['homophily'].std(),
        'Avg Positive Samples/Label': label_df['num_positive'].mean(),
        'Label Sparsity (%)': 100 * (1 - train_labels.sum() / (len(train_idx) * num_labels)),
    },
    'Feature Predictiveness': {
        'Top Feature Importance': feature_importance[top_features[0]],
        'Top 5 Features Avg': feature_importance[top_features[:5]].mean(),
        'Labels with Strong Correlation (>0.1)': (label_predictability > 0.1).sum(),
        'Max Label Predictability': label_predictability.max(),
        'Mean Label Predictability': label_predictability[label_predictability > 0].mean() if (label_predictability > 0).any() else 0,
    }
}

for category, values in metrics.items():
    print(f"\nðŸ“Š {category}:")
    for key, val in values.items():
        if isinstance(val, float):
            print(f"  â€¢ {key:.<50} {val:.4f}")
        else:
            print(f"  â€¢ {key:.<50} {val:,}" if isinstance(val, int) else f"  â€¢ {key:.<50} {val}")

print("\n" + "="*80)
print("DECISION RECOMMENDATIONS:")
print("="*80)

# Decision logic based on metrics
mean_homophily = label_df['homophily'].mean()
power_law_fit = r_value**2

print(f"\n1ï¸âƒ£ HOMOPHILY ANALYSIS (Current: {mean_homophily:.4f}):")
if mean_homophily > 0.6:
    print("  âœ… VERY HIGH â†’ Label Propagation will DOMINATE")
    print("  ðŸ“Š Recommended: LP weight = 70-80%, GAT weight = 20-30%")
    lp_weight_rec = "0.70-0.80"
elif mean_homophily > 0.4:
    print("  âœ… HIGH â†’ Label Propagation strongly recommended")
    print("  ðŸ“Š Recommended: LP weight = 60-70%, GAT weight = 30-40%")
    lp_weight_rec = "0.60-0.70"
elif mean_homophily > 0.25:
    print("  âš ï¸  MODERATE â†’ Balanced ensemble needed")
    print("  ðŸ“Š Recommended: LP weight = 50-60%, GAT weight = 40-50%")
    lp_weight_rec = "0.50-0.60"
else:
    print("  âš ï¸  LOW â†’ Feature-based models may be better")
    print("  ðŸ“Š Recommended: LP weight = 40-50%, GAT weight = 50-60%")
    lp_weight_rec = "0.40-0.50"

print(f"\n2ï¸âƒ£ NETWORK TOPOLOGY (Power-law fit RÂ² = {power_law_fit:.3f}):")
if power_law_fit > 0.8:
    print("  âœ… Strong scale-free structure â†’ Hub genes are critical")
    print("  ðŸ“Š Action: Use degree features (in/out/total) + attention mechanism (GAT)")
else:
    print("  âš ï¸  Weaker scale-free structure â†’ More uniform topology")
    print("  ðŸ“Š Action: Focus more on label propagation")

top_feat_importance = feature_importance[top_features[0]]
print(f"\n3ï¸âƒ£ FEATURE INFORMATIVENESS (Top feature: {top_feat_importance:.4f}):")
if top_feat_importance > 0.3:
    print("  âœ… Features are HIGHLY informative")
    print("  ðŸ“Š Action: GAT should learn good representations, increase GAT weight")
elif top_feat_importance > 0.15:
    print("  âœ… Features are MODERATELY informative")
    print("  ðŸ“Š Action: Balanced approach, use both LP and GAT")
else:
    print("  âš ï¸  Features are WEAKLY informative")
    print("  ðŸ“Š Action: Rely more on graph structure (LP), use degree features for GAT")

label_sparsity = 100 * (1 - train_labels.sum() / (len(train_idx) * num_labels))
print(f"\n4ï¸âƒ£ LABEL SPARSITY ({label_sparsity:.2f}%):")
if label_sparsity > 95:
    print("  âš ï¸  EXTREME sparsity â†’ Class imbalance is severe")
    print("  ðŸ“Š Action: MUST use weighted loss (pos_weight) + Correct & Smooth")
    print("  ðŸ“Š Action: Consider per-label thresholding")
else:
    print("  âœ… Moderate sparsity")
    print("  ðŸ“Š Action: Standard techniques should work")

print("\n" + "="*80)
print("ðŸŽ¯ FINAL RECOMMENDED STRATEGY:")
print("="*80)
print(f"\nâœ… Ensemble Configuration:")
print(f"  â€¢ Label Propagation + C&S: {lp_weight_rec}")
print(f"  â€¢ GAT + C&S: {1.0 - float(lp_weight_rec.split('-')[0]):.2f}-{1.0 - float(lp_weight_rec.split('-')[1]):.2f}")
print(f"\nâœ… GAT Configuration:")
print(f"  â€¢ Use degree features (+4 dimensions)")
print(f"  â€¢ 2-layer architecture with attention heads")
print(f"  â€¢ Weighted BCE loss with pos_weight={1/label_sparsity*100:.1f}")
print(f"\nâœ… Label Propagation:")
print(f"  â€¢ 50 iterations, alpha=0.9")
print(f"  â€¢ Exploits homophily={mean_homophily:.4f}")
print(f"\nâœ… Correct & Smooth:")
print(f"  â€¢ Apply to BOTH LP and GAT predictions")
print(f"  â€¢ 50 correction + 50 smoothing layers")
print(f"\nâœ… Expected Performance:")
if mean_homophily > 0.5 and power_law_fit > 0.8:
    print(f"  â€¢ Validation AP: 0.30-0.40 (optimistic)")
    print(f"  â€¢ Test AP: 0.25-0.35 (realistic)")
    print(f"  â€¢ ðŸŽ¯ Target >0.5 may require additional techniques")
elif mean_homophily > 0.3:
    print(f"  â€¢ Validation AP: 0.25-0.35")
    print(f"  â€¢ Test AP: 0.20-0.30")
    print(f"  â€¢ ðŸŽ¯ Target >0.5 will require advanced techniques")
else:
    print(f"  â€¢ Validation AP: 0.15-0.25")
    print(f"  â€¢ Test AP: 0.10-0.20")
    print(f"  â€¢ ðŸŽ¯ Target >0.5 unlikely with current approaches")

print("\nðŸ’¡ Additional Techniques to Consider:")
print("  1. Per-label thresholding (optimize threshold per disease)")
print("  2. Pseudo-labeling (use confident test predictions for retraining)")
print("  3. Multi-scale label propagation (different alpha values)")
print("  4. Ensemble of multiple LP runs with different seeds")
print("  5. Feature augmentation (centrality measures, community detection)")

print("\n" + "="*80)
```
